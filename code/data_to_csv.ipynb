```python
import os
import cv2
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path_arr = os.listdir("../data/image")
```

```python
image_path = []
mask_path = []
for i in range(0, len(path_arr)):
    image_path.append("../data/image/"+path_arr[i])
    mask_path.append("../data/mask/"+path_arr[i])

# intialise data of lists.
data = {'image_path':image_path,
        'mask_path':mask_path}
 
# Create DataFrame
df = pd.DataFrame(data)
 
# Print the output.
df.to_csv("data.csv", index=False)
```

```python
image = cv2.imread("../data/image/"+path_arr[0], cv2.IMREAD_UNCHANGED)
mask = cv2.imread("../data/mask/"+path_arr[0], cv2.IMREAD_UNCHANGED)
```
